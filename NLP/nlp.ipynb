## 2. Tokenization and Stopword Removal

```python
text = "It was a very pleasant day, the weather was cool and there were showers. I went to market to buy some fruits."
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sents = sent_tokenize(text)
```

```python
print(sents)
```

```python
words = word_tokenize(text)
```

```python
print(words)
```

```python
word_list = word_tokenize(sents[0].lower())
print(word_list)
```

### Removing Stop Words

```python
from nltk.corpus import stopwords
```

```python
sw = set(stopwords.words('english'))
print(sw, len(sw))
```

## Filter words from sentence

```python
def filter_words(word_list):
    useful_words =[ w for w in word_list if w not in sw]
    return useful_words
```

```python
useful_words = filter_words(word_list)
print(useful_words)
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
# tokenizer = RegexpTokenizer('[,.]')
tokenizer = RegexpTokenizer('[a-z0-9]+')
```

```python
sents = "send the 50 documents to abc, def, ghi."
print(tokenizer.tokenize(sents))
```

### Stemming 
- Process that transforms particular words into root words
- jumping, jump, jumps, jumped => jump

```python
text = " The quick brown fox was seen jumping over the lazy dog from high wall. Foxes love to make high jumps."
```

```python
word_list = tokenizer.tokenize(text.lower())
```

```python
print(word_list)
```

### Types of Stemmers
- Snowball Stemmer (Multilingual)
- Porter Stemmer
- Lancaster Stemmer

```python
from nltk.stem.snowball import PorterStemmer, SnowballStemmer
from nltk.stem.lancaster import LancasterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('jumped')
ps.stem('jumping')
```

```python
ps.stem('lovely')
```

```python
# ps.stem('awesome')
ls = LancasterStemmer()
ls.stem('awesome')
```

```python
print(ls.stem("teenager"))
print(ps.stem("teenager"))
```
